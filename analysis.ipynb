### Q.2
#### Identify and handle missing values column by column

```python
df.isnull().sum()
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent.sort_values(ascending=False)
```

```python
df["director"] = df["director"].fillna("Unknown")
```

```python
df["cast"] = df["cast"].fillna("Not Available")
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df["date_added"] = df["date_added"].fillna(df["date_added"].mode()[0])
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df = df.dropna(subset=["duration"])
```

### Q.3
#### Fix mixed-type columns (e.g., duration stored as "90 min")

```python
df[["duration_value", "duration_type"]] = df["duration"].str.split(" ", expand=True)
```

```python
df["duration_value"] = pd.to_numeric(df["duration_value"], errors="coerce")
```

```python
df["duration_type"] = df["duration_type"].replace({
    "min": "Minutes",
    "Season": "Seasons",
    "Seasons": "Seasons"
})
```

### Q.4
#### Parse date columns into proper datetime objects

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

#### Create as cleaned data in csv

```python
df.to_csv('cleaned-data.csv')
```
